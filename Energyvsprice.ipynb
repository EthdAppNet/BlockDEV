```python
# Initialize lists to store the results
import numpy as np
E_values = [50000]  # Initialize with the initial value of E
p_values = []

# Initial values
E = 50000
L = 50000

# Generate 100 different random values between 200 and 10,000
#delta_e_values = random.sample(range(5, 200), 150)
delta_e_values =  [random.uniform(500, 2000) for _ in range(200)]
# Function to format the x-axis labels with thousands separator
def format_thousands(x, pos):
    return f'{int(x/1000)}K'

# Initialize labels for the x-axis
x_labels = [E]

# Store the values of available energy at specific indices
price_change = []

# Loop through the values of Delta_e randomly
for i, Delta_e in enumerate(delta_e_values):
    # Decide randomly whether to add or subtract Delta_e
    add_delta_e = random.choice([True, False])

    if add_delta_e:
        # Calculate p for adding Delta_e
        delta_m = L * Delta_e / (E + Delta_e)
        E = E + Delta_e
        L = L - delta_m
        P_new = L / E
    else:
        # Calculate p for subtracting Delta_e
        delta_m = L * Delta_e / (E - Delta_e)
        E = E - Delta_e
        L = L + delta_m
        P_new = L / E

    # Append the calculated E and P_new to their respective lists
    E_values.append(E)
    p_values.append(P_new)
    x_labels.append(E)
    # Store the values of available energy at specific indices (0, 50, 100, 150)
    if i in [0, 50, 100, 150]:
        price_change.append(E / 50000)


# Create a list of indices for the x-axis ticks
x_ticks_indices = list(range(1, len(x_labels)))

# Create a figure with two y-axes on the left and right
fig, ax1 = plt.subplots(figsize=(10, 6))

# Plotting the results for p on the left y-axis without markers and add legend
ax1.plot(x_ticks_indices, p_values, linestyle='-', color='tab:blue', label='Price/Kwh ($)')
ax1.set_xlabel('Transactions')
ax1.set_ylabel('Price/Kwh ($)', color='tab:blue')
ax1.tick_params(axis='y', labelcolor='tab:blue')

# Create a second y-axis on the right for E and add legend
ax2 = ax1.twinx()
ax2.plot(x_ticks_indices, E_values[1:], linestyle='-', color='tab:red', label='Available Energy (Kwh)')
ax2.set_ylabel('Energy (Kwh)', color='tab:red')
ax2.tick_params(axis='y', labelcolor='tab:red')

# Combine legends from both axes
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
lines = lines1 + lines2
labels = labels1 + labels2
ax1.legend(lines, labels, loc='upper left')

# Set custom x-axis labels
#plt.xticks(x_ticks_indices, x_ticks_indices)

# Format the x-axis labels with thousands separator
# ax1.xaxis.set_major_formatter(plt.FuncFormatter(format_thousands))
plt.grid(True, which='both')
plt.savefig('price(500-2000).jpg', dpi=300, bbox_inches='tight')

plt.show()
```
