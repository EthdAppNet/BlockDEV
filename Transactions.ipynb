```python
import matplotlib.pyplot as plt

# Number of CSs on the x-axis
cs_values = list(range(1, 10))

# Calculate the number of transactions for the proposed model and baseline
proposed_model_transactions = [  cs +2 for cs in cs_values]
baseline_transactions = [(2 * cs) +3 for cs in cs_values]  # Scaling the baseline line

# Create the graph
plt.figure(figsize=(10, 6))
plt.plot(cs_values, proposed_model_transactions, label="BlockDEV")
plt.plot(cs_values, baseline_transactions, label="Baseline")

# Add labels and a legend
plt.xlabel("Number of CSs")
plt.ylabel("Number of Transactions")
plt.legend()
plt.grid()
# Show the graph
plt.savefig('CSTransactions.jpg', dpi=300, bbox_inches='tight')
plt.show()

from google.colab import files
files.download('CSTransactions.jpg')
```

```python
import matplotlib.pyplot as plt

# Number of CSs on the x-axis
ev_values = list(range(1, 10))

# Calculate the number of transactions for the proposed model and baseline
proposed_model_transactions = [2 * ev + 1 for ev in ev_values]
baseline_transactions = [3 * ev +2 for ev in ev_values]  # Scaling the baseline line

# Create the graph
plt.figure(figsize=(10, 6))
plt.plot(ev_values, proposed_model_transactions, label="BlockDEV")
plt.plot(ev_values, baseline_transactions, label="Baseline")

# Add labels and a legend
plt.xlabel("Number of EVs")
plt.ylabel("Number of Transactions")
plt.legend()
plt.grid()
# Show the graph
plt.savefig('EVTransactions.jpg', dpi=300, bbox_inches='tight')
plt.show()

from google.colab import files
files.download('EVTransactions.jpg')
```
