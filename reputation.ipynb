```python
import numpy as np
import matplotlib.pyplot as plt

# Define the parameters
gamma = 1  # You can change this value to suit your needs
alpha = 1  # You can change this value to suit your needs
beta = 1
n = 1  # Fixed value of n
#Random delta values 0 to 100%
delta_d_values = np.random.uniform(0, 10, 40)  # Range of ∆D values

# Initialize arrays to store the values
Pc1_values = np.zeros(len(delta_d_values))
Pc_values = np.zeros(len(delta_d_values))

# Calculate Pc1 for a fixed n and ∆D
for i, delta_d in enumerate(delta_d_values):
    if n == 0:
        Pc1_values[i] = 0  # Initial condition
    else:
        Pc1_values[i] = Pc1_values[i - 1] + np.exp(-beta * delta_d)

# Calculate Pc and P for a fixed n and ∆D
for i, delta_d in enumerate(delta_d_values):
    if n == 0:
        Pc_values[i] = 0  # Initial condition
    else:
        Pc_values[i] = 1 / (1 + gamma * np.exp(-alpha * Pc1_values[i]))

# Reverse the arrays for plotting
Pc1_values_reversed = np.flip(Pc1_values)
Pc_values_reversed = np.flip(Pc_values)

# Plot the graph
plt.plot(Pc1_values_reversed, Pc_values_reversed, label='Our model')
###############################

y_values = np.arange(0.5, 1.2, 0.1)

# Generate corresponding x values (assuming indices as x values)
x_values = np.arange(len(y_values))
print (x_values)

# Plot the line
plt.plot(x_values, y_values, label='Baseline')

######################################################

# Add a horizontal line at y-axis value 1
plt.axhline(y=1, color='gray', linestyle='--')


plt.xlabel('Number of charging sessions (n)')
plt.ylabel(r'$P^n_r$')
#plt.title(f'Pc vs. Pc^n_1 for n = {n}')
plt.legend()
plt.grid(True)

# Remove x-axis ticks and add an arrow annotation
plt.xticks([])
plt.annotate('---->', xy=(len(Pc1_values_reversed) - 1, Pc_values_reversed[-1]), xytext=(len(Pc1_values_reversed) - 2, Pc_values_reversed[-1]),
             arrowprops=dict(facecolor='black', arrowstyle='->'))
plt.savefig('reputation.jpg', dpi=300, bbox_inches='tight')
plt.show()
```
