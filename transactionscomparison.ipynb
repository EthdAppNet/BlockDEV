```python
import matplotlib.pyplot as plt

# Data
categories = ['BlockDEV', 'Baseline']
values = [3, 5]
colors = ['#1f77b4', '#ff7f0e']
# Create a bar plot


plt.figure(figsize=(10, 6))
plt.bar(categories, values, color=colors,  width=0.1)


plt.ylabel('Blockchain transactions per flow')


# Display the plot

plt.grid(True)
plt.savefig('trans.jpg', dpi=300, bbox_inches='tight')
plt.show()
from google.colab import files
files.download('trans.jpg')
```
